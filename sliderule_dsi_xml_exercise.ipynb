```python
document = ET.parse( './data/mondial_database.xml' )
```

```python
for child in document.getroot():
    print child.tag
```

```python
#10 countries with the lowest infant mortality rates
import pandas as pd

x=[]
y=[]
for element in document.iterfind('country'):
    if element.find('infant_mortality') is not None:
        #print  element.find('name').text + '     ' + element.find('infant_mortality').text
        x.append(str(element.find('name').text))
        y.append(float(element.find('infant_mortality').text))
#print x,y 
df = pd.DataFrame({'Country': x,'Infant Mortality':y})
#print df
df.sort_values('Infant Mortality').head(10)
    
```

```python
#10 cities with the largest population
import pandas as pd

countryList=[]
cityList=[]
maxPopList=[]
for element in document.iterfind('country'): 
    for subelement in element.getiterator('city'):
        if subelement.find('population') is not None:
            arrPop=[] #list to hold the populations over years for each city
            for pop in subelement.findall('population'): 
                arrPop.append(float(pop.text))
                #print subelement.find('name').text, pop.attrib, pop.get('year'), pop.text
            #print arrPop
            countryList.append(element.find('name').text)
            cityList.append(subelement.find('name').text)
            maxPopList.append(max(arrPop)) #saves largest recorded population or should it be most recent (2011)?? 
            
df = pd.DataFrame({'Country': countryList,'City': cityList, 'Pop' : maxPopList})
#print df
df.sort_values('Pop', ascending=False).head(10)
```

```python
#name and country of a) longest river, b) largest lake and c) airport at highest elevation

#lake script prints country bordering longest river + length
riverLen = 0.0
cntryName = ""
riverName = ""
for river in document.iterfind('river'):
    if ((river.find('length') is not None) and float(river.find('length').text) > riverLen):
        riverLen = float(river.find('length').text)
        cntryName = str(river.get('country'))
        riverName = str(river.find('name').text)
#gets the names of the countries listed in country name string
cntryArr = []
for name in cntryName.split() : 
    for element in document.iterfind('country'):
        tmp_str=str(element.get('car_code'))
        if (tmp_str==name):
            cntryArr.append(element.find('name').text)
            break
print 'Country: ' + str(cntryArr) + ', River: ' + riverName + ', Length: ' + str(riverLen)


#lake script prints country bordering largest lake + area
lakeSz = 0.0
cntryName2 = ""
lakeName = ""
for lake in document.iterfind('lake'):
    if ((lake.find('area') is not None) and float(lake.find('area').text) > lakeSz):
        lakeSz = float(lake.find('area').text)
        cntryName2 = str(lake.get('country'))
        lakeName = str(lake.find('name').text)

cntryArr = []
for name in cntryName2.split() : 
    for element in document.iterfind('country'):
        tmp_str=str(element.get('car_code'))
        if (tmp_str==name):
            #print element.find('name').text, element.get('car_code')
            cntryArr.append(element.find('name').text)
            break
    #print cntryArr
print 'Country: ' + str(cntryArr) + ', Lake: ' + lakeName + ', Area: ' + str(lakeSz)


        
        
        
#airport elevation script prints country with highest airport + elevation
apElev = 0.0
tmp_apElev = 0.0
cntryName3 = ""
apName = ""
#aplist = [] for checking/debugging
for ap in document.iterfind('airport'):
    #print ap.find('name').text, ap.attrib, ap.find('elevation').text, ap.get('country'), apElev    
    if (ap.find('elevation').text is not None):
        tmp_apElev = float(ap.find('elevation').text) #had to create this tmp var to get around wierd type issue
        #aplist.append(tmp_apElev)
        if apElev < tmp_apElev : 
            apElev = tmp_apElev
            cntryName3 = str(ap.get('country'))
            apName = ap.find('name').text
            
cntryArr = []
for name in cntryName3.split() : 
    for element in document.iterfind('country'):
        tmp_str=str(element.get('car_code'))
        if (tmp_str==name):
            cntryArr.append(element.find('name').text)
            break
print 'Country: ' + str(cntryArr) + ', Airport: ' + apName + ', Elevation: ' + str(apElev)
```
